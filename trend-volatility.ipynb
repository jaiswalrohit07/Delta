```python
import talib
```

```python
df['sar'] = talib.SAR(df['high'], df['low'], acceleration=0.02, maximum=0.2)
```

```python
df.head()
```

```python
df['position'] = np.where((df['close'] > df['sar']) & (df['close'].shift(1) < df['sar'].shift(1)) , 1, 0)
df['position'] = np.where((df['close'] < df['sar']) & (df['close'].shift(1) > df['sar'].shift(1)) , -1, df['position'])
```
